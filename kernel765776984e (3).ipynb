```python
train_path="/kaggle/input/chest-xray-pneumonia/chest_xray/chest_xray/train";
test_path="/kaggle/input/chest-xray-pneumonia/chest_xray/chest_xray/test";
val_path="/kaggle/input/chest-xray-pneumonia/chest_xray/val";
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
import PIL
a=tf.keras.preprocessing.image.load_img('/kaggle/input/chest-xray-pneumonia/chest_xray/chest_xray/train/PNEUMONIA/person373_virus_756.jpeg')
plt.imshow(a)
b=tf.keras.preprocessing.image.img_to_array(a)
import numpy as np
c=np.expand_dims(b,axis=0).shape
print(c)
```

```python
train_datagen=tf.keras.preprocessing.image.ImageDataGenerator(rescale=1.0/255)
test_datagen=tf.keras.preprocessing.image.ImageDataGenerator(rescale=1.0/255)
val_datagen=tf.keras.preprocessing.image.ImageDataGenerator(rescale=1.0/255)
```

```python
training_set=train_datagen.flow_from_directory(train_path,target_size=(128,128),batch_size=5216,class_mode='binary')
test_set=test_datagen.flow_from_directory(test_path,target_size=(128,128),batch_size=624,class_mode='binary')
val_set=test_datagen.flow_from_directory(val_path,target_size=(128,128),batch_size=16,class_mode='binary')
```

```python
x1,y1=training_set.next()
x2,y2=test_set.next()
x3,y3=val_set.next()
```

```python
x1.shape
```

```python
x2.shape
```

```python
x3.shape
```

```python
x=np.concatenate((x1,x2),axis=0)
x=np.concatenate((x,x3),axis=0)
```

```python
y=np.concatenate((y1,y2),axis=0)
y=np.concatenate((y,y3),axis=0)
```

```python
x.shape
```

```python
x_train=x[:5000]
x_test=x[5000:5600]
x_val=x[5600:]
y_train=y[:5000]
y_test=y[5000:5600]
y_val=y[5600:]
```

```python
train_datagen=tf.keras.preprocessing.image.ImageDataGenerator(shear_range=0.3,zoom_range=0.3,horizontal_flip=True)
test_datagen=tf.keras.preprocessing.image.ImageDataGenerator()
val_datagen=tf.keras.preprocessing.image.ImageDataGenerator()
```

```python
training_data=train_datagen.flow(x_train,y_train,batch_size=64)
```

```python
model=tf.keras.models.Sequential();
model.add(tf.keras.layers.Conv2D(32,(3,3),input_shape=(64,64,3),activation=tf.nn.relu))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
model.add(tf.keras.layers.Conv2D(40,(3,3),activation=tf.nn.relu))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
model.add(tf.keras.layers.Conv2D(50,(3,3),activation=tf.nn.relu))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(units=128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(units=64,activation=tf.nn.relu))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(units=32,activation=tf.nn.relu))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(units=1,activation=tf.nn.sigmoid))
```

```python
vgg16=tf.keras.applications.VGG16(
    include_top=False,
    weights="imagenet",
    input_shape=(128,128,3),
    pooling='avg'
)
```

```python
inception=tf.keras.applications.ResNet50(
    include_top=False,
    weights="imagenet",
    input_shape=(128,128,3),
    pooling='max'
) 
```

```python
model=tf.keras.models.Sequential([
    vgg16,
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
model.layers[0].trainable=False
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit_generator(training_data,epochs=25,steps_per_epoch=5000/64 ,validation_data=(x_val,y_val))
```
